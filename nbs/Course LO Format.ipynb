```python
import re
def generate_display_fields(df):
    standardized_set = []
    for idx, i in enumerate(df.activity):
        try:
            shortened_str = re.sub("[^0-9,.-]", "", i)
            sst = shortened_str.split(".")
            if len(sst) == 2:
                standardized = 'U' + sst[0] + 'S' + sst[1]
            else:
                standardized = "U" + sst[0] + "S" + sst[1] + "T" + "".join(sst[2:])

            if standardized[:2] == 'U-':
                standardized = "U" + standardized[2:]
        except Exception as e:
            print(e, idx)
            standardized = ""

        standardized_set.append(standardized)
        
    return standardized_set

def generate_activity_type(df):
    activity_set = []
    option_set = ['ALM', 'GPSM', 'PSM', None]
    for idx, i in enumerate(df.activity):
        c = -1
        if option_set[0] in i:
            c = 0
        elif option_set[1] in i:
            c = 1
        elif option_set[2] in i:
            c = 2
        else:
            c = 3
        activity_set.append(option_set[c])
    return activity_set

def generate_duration_cols(df):
    dur_removed, dur = [], []
    word, length = 'duration', 10
    for idx, i in enumerate(df.lo_formatted):
        try:
            assert len(dur) == len(dur_removed)
        except:
            print(idx, df.lo_formatted[idx-1])
            assert(0)
        if word in i:
            ind = i.index(word)
            dur_removed.append(i[ind + length:])
            sst = i[:ind].split()
            c = 0
            for j in sst:
                if j.isnumeric():
                    c = 1
                    dur.append(j)        
            if not c:
                dur.append(0)
        else:
            dur_removed.append(i)
            dur.append(0)
    return dur_removed, dur
        
def generate_ust_cols(df):
    s_set = []
    for i in list(df['activity_formatted']):
        q = tuple([c for c in re.split(r'(\d+)', i) if c.isdigit()])
        s_set.append(q)
    return s_set
```

```python
output_folder = "../output/los_course_output/"
import numpy as np
def generate_course_files(df, out_path):
    courses = set(df['course'])
    for course in tqdm(courses):
        sub_df = df[df.course == course]
        try:
            sorted_sub_df = sub_df.sort_values(by='sorting_col')
            ss_df_copy = sorted_sub_df.copy()
            ss_df_copy = ss_df_copy.drop(columns=['sorting_col'])
            print(f"Outputting csv for course {course}")
            csv_path = out_path + course + ".csv"
            ss_df_copy.to_csv(csv_path, index=False)
        except Exception as e:
            print(f"ERROR for course {course}, {e}")
            
subset_df = df.iloc[indices] # Create subset dataframe        
generate_course_files(subset_df, output_folder)
```

```python
import re

def atoi(text):
    return int(text) if text.isdigit() else text

def helper(texts):
    b = list(texts).copy()
    b.sort(key=natural_keys)
    return b

def natural_keys(text):
    '''
    alist.sort(key=natural_keys) sorts in human order
    http://nedbatchelder.com/blog/200712/human_sorting.html
    (See Toothy's implementation in the comments)
    '''
    
    return [ atoi(c) for c in re.split(r'(\d+)', text) ]

subset_df.sort_values(by='activity_formatted_topic', key=helper)
```
