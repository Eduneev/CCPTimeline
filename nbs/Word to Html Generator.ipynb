```python
import os
import shutil

def generate_html(input_file, output_file):
    # Apply the mammoth conversion and write the HTML output to the output file
    with open(input_file, "rb") as docx_file:
        result = mammoth.convert_to_html(docx_file)
        html = result.value  # The generated HTML
        print(f"Outputting converted contents of {input_file} to {output_file}")
        with open(output_file, "w", encoding="utf-8") as html_file:
            html_file.write(html)
        messages = result.messages
        #print(f"Messages: {messages}")

def process_files(input_path, output_path, process_function):
    # Create the output folder if it doesn't exist
    os.makedirs(output_path, exist_ok=True)

    # Iterate over files and subdirectories in the input folder
    for entry in os.scandir(input_path):
        if entry.is_file() and entry.name.endswith('.docx'):
            # Process file
            #print(entry.name)
            #print(os.path.splitext(entry.name)[0] + '.html')
            #continue
            
            output_file_name = os.path.splitext(entry.name)[0] + '.html'
            output_file = os.path.join(output_path, output_file_name)
            process_function(entry.path, output_file)
        elif entry.is_dir():
            # Recursively process subfolder
            subfolder_output = os.path.join(output_path, entry.name)
            process_files(entry.path, subfolder_output, process_function)


# Example usage: specify input and output paths
input_folder = '../data/unit_notes/'
output_folder = '../output/unit_notes/'

process_files(input_folder, output_folder, generate_html)
```
